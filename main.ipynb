```python
import numpy as np
```

```python
dictionary_txt = open("vendors/dictionary/phone.cin.txt", "r",encoding="utf-8")
```

```python
dictionary_list = list()
dictionary_list_index = list()
```

```python
for i, line in enumerate(dictionary_txt):
    if list(line)[0] is "%":
        continue
        
    line_content = [i.replace("\n", "") for i in line.replace("  ", " ").split(" ")]
    dictionary_list.append(line_content)
    dictionary_list_index.append([
        list(reversed(list(line_content[0]))),
        line_content[1]
    ])
```

```python
def find_close_word_with_voice(find, match_level):
    close_data = list()
    find_index = list(reversed(list(find)))
    
    for data in dictionary_list_index:

        match_time = 0

        try:
            for index, key in enumerate(find_index):
                if len(data[0]) > index:
                    if data[0][index] == key:
                        match_time += 1
                    else:
                        break
        except:
            print(data)

        if match_time > (match_level - 1):
            close_data.append([match_time, data])

    return close_data
```

```python
def find_word_key(word):
    for data in dictionary_list:
        if data[1] == word:
            return data[0]
        
    return "error"
```

```python
find_close_word_with_voice("cl3", 3)
```

```python
find_close_word_with_voice("ul3", 3)
```

```python
find_word_key("眇")
```

```python
def find_close_word(word, match_level):
    return find_close_word_with_voice(find_word_key(word), match_level)
```
